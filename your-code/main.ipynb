## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
import math as m

W = 10          # White balls
R = 12          # Red balls
B = 8           # Black balls

total_outcome = W + R + B

# Probability of taking a white ball out
p_white = W / total_outcome
print(f"White ball probability: {p_white:.2f}")

# White first, black second (without replacement)
p_white_first = W / total_outcome
p_black_second = B / (total_outcome - 1)

p_white_black_no_replace = p_white_first * p_black_second
print(f"White then black (no replacement): {p_white_black_no_replace:.2f}")

# White first, black second (with replacement)
p_white_first = W / total_outcome
p_black_second = B / total_outcome

p_white_black_replace = p_white_first * p_black_second
print(f"White then black (with replacement): {p_white_black_replace:.2f}")

# 4. Taking 3 balls and all being different colours (no replacement)
# One white, one red, one black in any order

total =  m.comb(total_outcome, 3)
p_white_comb = m.comb(W, 1)
p_red_comb = m.comb(R, 1)
p_black_comb = m.comb(B, 1)
fav_outcomes = p_black_comb* p_red_comb* p_white_comb
p_add_diff_color = fav_outcomes/total

print(f"3 balls all different colours: {p_add_diff_color:.2f}")

```

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

```python
# YOUR CODE HERE

# Total_week_day = 7

def saturday_rain_probability() -> float:

    # 40% are cloudy days
    cloudy_days = 0.40 * 7

    non_cloudy_days = 7 - cloudy_days

    # 50% are cloudy days are rainy days
    rainy_cloudy_days = 0.50 * cloudy_days
    
    non_rainy_cloudy_days = 0.2 * non_cloudy_days

    # total_rainy days in week
    total_rainy_days = rainy_cloudy_days + non_rainy_cloudy_days

    # Probabibity that saturday is rainy
    p_saturday_rainy = total_rainy_days / 7 * 100

    return p_saturday_rainy

print(f"Probability that saturday is rainy: {saturday_rain_probability():.1f}%")
```

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

```python
# YOUR CODE HERE

# Given probabilities
p_A = 0.3    # Probability of event A
p_B = 0.6    # Probability of event B
p_AB = 0.1   # Probability of A and B occurring together

# ---Calculate the needed probabilities---

# Probability of NOT A
p_not_A = 1 - p_A

# Probability of NOT B
p_not_B = 1 - p_B

# Probability of A OR B (union formula)
p_A_union_B = p_A + p_B - p_AB

# Probability of A AND NOT B (A minus intersection)
p_A_and_not_B = p_A - p_AB

# Probability of NOT A AND NOT B (outside both circles)
p_not_A_and_not_B = 1 - p_A_union_B

# Probability of NOT A AND B (B minus intersection)
p_not_A_and_B = p_B - p_AB

# ---Check each statement---

# Statement 1: P(Ā) = 0.7
print("Statement 1:", p_not_A == 0.7)

# Statement 2: P(A ∪ B) = 0.6
print("Statement 2:", p_A_union_B == 0.6)

# Statement 3: P(A ∩ B̄) = 0.2
print("Statement 3:", p_A_and_not_B == 0.2)

# Statement 4: P(A ∪ Ā) = 1 (always true)
print("Statement 4:", (p_A + p_not_A) == 1)

# Statement 5: P(A ∩ (B ∪ B̄)) = 0.4
# Note: B ∪ B̄ = sample space, so this equals P(A)
print("Statement 5:", p_A == 0.4)

# Statement 6: P(Ā ∩ B̄) = 0.3
print("Statement 6:", p_not_A_and_not_B == 0.3)

# Statement 7: P(Ā ∩ B) = 0.9
print("Statement 7:", p_not_A_and_B == 0.9)
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# Probability of getting COVID if flying to China
p_covid_given_china = 0.01

# Probability of getting COVID if flying to Italy
p_covid_given_italy = 0.20

# Probability of flying to Italy
p_italy_trip = 1 / 6

# Probability of flying to China
p_china_trip = 5 / 6

# Probability of flying to China AND getting COVID
p_china_and_covid = p_china_trip * p_covid_given_china

# Probability of flying to Italy AND getting COVID
p_italy_and_covid = p_italy_trip * p_covid_given_italy

# Total probability of getting COVID, either flying to Italy and China
p_covid = p_china_and_covid + p_italy_and_covid * 100
print(f"Probability of getting COVID: {p_covid:.1f}%")

# Conditional probability: P(China | COVID)
p_china_given_covid = (p_china_and_covid / p_covid) * 100
print(f"Probability that the trip was to China given COVID: {p_china_given_covid:.1f}%")
```
